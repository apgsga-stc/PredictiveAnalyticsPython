# Imports

```python
from UpliftTarget import Variable, And, Or
```

# Functions

```python
def best_slots(target, column, top_n: int = 20, where=""):
    if where:
        data = target.result.query(where)
    else:
        data = target.result
    return data.sort_values(column, ascending=False).head(top_n)
```

# Target Group Definitions

## Automotive

```python
TargetAuto = And(
    "Automotive",
    Or(
        "Auto nicht billig",
        Variable(name="Teures Auto", variable="g_CarHigh", code_nr=[1]),
        Variable(name="Mittelteures Auto", variable="g_CarMedium", code_nr=[1]),
    ),
    Variable(name="Autobesitz", variable="g_220", code_nr=[1, 2]),
)

print(TargetAuto.description())
```

```python
TargetAuto.set_timescale("Hour")
```

```python
%%time
TargetAuto.calculate()
```

```python
TargetAuto.export_result()
```

### Best Slots

```python
best_slots(TargetAuto, "pop_uplift", top_n=20, where="target_pers > 1000")
```
